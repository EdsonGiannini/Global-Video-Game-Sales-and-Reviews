# Importação das Bibliotecas Necessárias e Carga de Dados

Esta seção consolida todas as importações de bibliotecas necessárias para a análise, visualização e pré-processamento dos dados, bem como o carregamento inicial do dataset Iris.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
```

# Limpeza e pré-processamento de Dados

```python
#Retidrando colunas não usadas
del df['index']
del df['Rank']
del df['Game Title']
del df['North America']
del df['Europe']
del df['Japan']
del df['Rest of World']

#Normalizando os números
df = df.round(2)
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')
```

# Lidando com valores Faltantes

```python
#Procurando linhas com dados faltantes
df[df.isnull().any(axis=1)]
```

**Como a quantidade de linhas com algum dado faltante era menor do que 2% resolvi deleta-las.**
>


```python
df.dropna(subset=['Year'], inplace=True)
df.dropna(subset=['Publisher'], inplace=True)
```

```python
#Vizualização da tabela formatada
df
```

Analisaremos agora a média de reviews por ano para descobrir o ano com as melhores notas


```python
# média dos atributos numéricos do dataset
df[['Global', 'Review']].describe().loc['mean']
```

```python
medias = df[['Year', 'Global', 'Review']].groupby('Year').mean()

medias_review = df.groupby('Year')['Review'].mean().reset_index()

# Gráfico de Linhas
plt.figure(figsize=(20, 5))
plt.plot(
    medias_review['Year'],
    medias_review['Review'],
    marker='o',
    color='#e4000f',
    linestyle='-',
    linewidth=2,
    markersize=8
)

plt.title('Média de Reviews por Ano', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Média Review', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(medias_review['Year'])  # Mostrar todos os anos no eixo X
plt.show()
```

## Hipótese 3

```python
contagem_plataformas = df['Platform'].value_counts()

print(contagem_plataformas)
```

```python
mapeamento_fabricantes = {
    'PS2': 'Sony Computer Entertainment',
    'PS': 'Sony Computer Entertainment',
    'PS3': 'Sony Computer Entertainment',
    'PSP': 'Sony Computer Entertainment',
    'PSV': 'Sony Computer Entertainment',
    'X360': 'Microsoft Game Studios',
    'XB': 'Microsoft Game Studios',
    'Wii': 'Nintendo',
    'DS': 'Nintendo',
    'GBA': 'Nintendo',
    'NES': 'Nintendo',
    'N64': 'Nintendo',
    'GC': 'Nintendo',
    'GB': 'Nintendo',
    'SNES': 'Nintendo',
    '3DS': 'Nintendo',
    'WiiU': 'Nintendo',
    'GEN': 'Sega',
    'DC': 'Sega',
    'SAT': 'Sega',
    'SCD': 'Sega',
    'PC': '"not specific"',
}

df['Platform Owner'] = df['Platform'].map(mapeamento_fabricantes)
```

```python
df['Platform Owner'].value_counts()
```

```python
filtro = df[df['Publisher'] == df['Platform Owner']]
print(filtro)
```

```python
media_vendas = filtro.groupby(['Publisher', 'Platform Owner'])['Global'].mean().reset_index()
print(media_vendas)
```

```python
filtro = df[df['Publisher'] == df['Platform Owner']]

media_vendas = filtro.groupby(['Publisher', 'Platform Owner'])['Global'].mean().reset_index()

media_geral = df['Global'].mean()

plt.figure(figsize=(10, 6))
bars = plt.bar(
    media_vendas['Publisher'],
    media_vendas['Global'],
    color=['#0e7a0d','#e4000f', '#17569b', 'black'],
    label='Média por Editora'
)

# Adicionar linha da média geral
plt.axhline(
    y=media_geral,
    color='orange',
    linestyle='--',
    linewidth=2,
    label=f'Média Geral: {media_geral:.1f}'
)

# Personalização
plt.title('Média de vendas quando a dona do console é a editora', fontsize=14)
plt.xlabel('Editora', fontsize=12)
plt.ylabel('Média de Vendas', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()  # Mostrar legenda
plt.show()
```

```python
media_reviews = filtro.groupby(['Publisher', 'Platform Owner'])['Review'].mean().reset_index()
print(media_reviews)
```

```python
filtro = df[df['Publisher'] == df['Platform Owner']]

media_vendas = filtro.groupby(['Publisher', 'Platform Owner'])['Review'].mean().reset_index()

media_geral = df['Review'].mean()

plt.figure(figsize=(10, 10))
bars = plt.bar(
    media_vendas['Publisher'],
    media_vendas['Review'],
    color=['#0e7a0d','#e4000f', '#17569b', 'black'],
    label='Média por Editora'
)

# Adicionar linha da média geral
plt.axhline(
    y=media_geral,
    color='orange',
    linestyle='--',
    linewidth=2,
    label=f'Média Geral: {media_geral:.1f}'
)

# Personalização
plt.title('Média de Reviews quando a dona do console é a editora', fontsize=14)
plt.xlabel('Editora', fontsize=12)
plt.ylabel('Média de Vendas', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()  # Mostrar legenda
plt.show()
```
